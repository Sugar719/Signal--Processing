```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from matplotlib import pyplot
import matplotlib

plt.rcParams['font.sans-serif']=['PingFang HK']
```

# 任务一

```python
Fs = 200000.0
Ts = 2.0/Fs
t = np.arange(-1,1,Ts)
fs1 = 10

y1 = 2*(np.sin(2*np.pi*fs1*t)) #sin信号
plt.title(u'10Hz的正弦信号')
plt.plot(t, y1)
plt.xlim(-0.5, 0.5)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

```python
fs2 = 20
phi = (50/180) * np.pi
y2 = 4*(np.cos(2*np.pi*fs2*t + phi))#cos信号
# plt.axhline(y=0,ls=":",c="red")
plt.title(u'20Hz的余弦信号')
plt.plot(t, y2)
plt.xlim(-0.5, 0.5)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

```python
# plt.axhline(y=0,ls=":",c="red")
plt.title(u'相加信号')
plt.plot(t,y1+y2)
plt.xlim(-0.5,0.5)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

```python
X = np.arange(-20, 20, 0.01)
Y = []
def f(x):
    return np.sin(x)/x
for i in X:
    v, err = integrate.quad(f, 0, i, points=[0])
    Y.append(v)
# plt.axhline(y=0,ls=":",c="red")
plt.title(u'Si(t)信号')
plt.plot(X,Y)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

```python
def fun(x,x1,x2):
    if x < x1:
        return x-1
    elif x >= x2:
        return x-2
    else:
        return 1
X = np.linspace(1, 4, 3000)
Y = np.array([fun(t, 2, 3) for t in X])
plt.title(u'f(t)')
plt.plot(X,Y)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

```python
def fun(x,x1,x2):
    if x < x1:
        return (-2)*x+1
    elif x >= x2:
        return (-2)*x+2
    else:
        return 1
X = np.linspace(-1/2, 1, 1000)
Y = np.array([fun(t, 0, 1/2) for t in X])
plt.plot(X,Y)
plt.title(u'f(3-2t)')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

T0 = 0.5

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

N1 = 20000
t = np.linspace(-1, 1, N1, endpoint=False)
sq = signal.square(2 * np.pi * 2 * t) #square函数生成方波

# 方波信号
plt.plot(t, sq)
plt.xlim(-0.75, 0.75)
plt.title(u'方波信号-4Hz 采样10000Hz')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.ylabel('Amplitude', fontdict={'weight':'normal','size': 15})
plt.xlabel('Time', fontdict={'weight':'normal','size': 15})
plt.show()

n = [2*i+1 for i in range(10)]
w = []
A_n = []
phi = []
for i in n:
    an = 4/math.pi * (1/i)
    A_n.append(an)
    wn = 4*math.pi * i
    w.append(wn)
    phi.append(0)

# 三角型频谱图
plt.stem(w,A_n, label="all")
plt.xlim(0, 100)
plt.axhline(y=0,ls="-",c="red")
plt.title(u'方波信号-4Hz 三角型频谱')
plt.ylabel(u'An', fontdict={'weight':'normal','size': 15})
plt.xlabel(u'ω', fontdict={'weight':'normal','size': 15})
plt.show()

# 复数型频谱图
plt.stem(w,phi, label="all")
plt.xlim(0, 100)
plt.axhline(y=0,ls="-",c="red")
plt.title(u'方波信号-4Hz 复数型频谱')
plt.ylabel(u'φ', fontdict={'weight':'normal','size': 15})
plt.xlabel(u'ω', fontdict={'weight':'normal','size': 15})
plt.ylim(-0.1,1.2)
plt.show()
```
