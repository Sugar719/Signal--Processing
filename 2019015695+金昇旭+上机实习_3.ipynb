```python
from tkinter import font
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist

t = symbols('t')
y=Symbol('y')
f=Symbol('f')
n=Symbol('n')
#频截为50hz，频谱为三角
fm=50
X=Piecewise(((1-abs(f/fm)),abs(f)<fm), (0, True))
X
```

```python
v=np.arange(-100,100,0.1)
z=[X.subs(f,i) for i in v]
plt.plot(v,z)
plt.title('频谱为三角型、最高截频为50Hz的带限连续信号')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
# plt.savefig("D:\桌面\\1(1).jpg",dpi=300)
plt.show()
```

```python
x=integrate(X*exp(2j*pi*f*t),(f,-fm,fm))
v=np.arange(-0.1,0.1+1e-10,1/500)
v[abs(v)<1e-10]=0
z=[complex(x.subs(t,i).evalf()) for i in v]
plt.plot(v,z)
plt.xlim(-0.11,0.11)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.title('时间域')
# plt.savefig("D:\桌面\\1(2).jpg",dpi=300)
plt.show()
```

```python
v=np.linspace(-0.1,0.1+1e-10,num=11)
v[abs(v)<1e-10]=0
z=[complex(x.subs(t,i).evalf()) for i in v]
plt.stem(v,z)
plt.plot(v,z)
#设置为红色线条
plt.xlim(-0.11,0.11)
plt.title('采样-50Hz')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
# plt.savefig("D:\桌面\\1(3).jpg",dpi=300)
plt.show()
```

```python
delta=1/50
m=symbols('m')
f=Symbol('f')
X_delta=summation(X.subs(f,f+m/delta),(m,-50,50))
v=np.arange(-300,310,10)
z=[X_delta.subs(f,i).evalf() for i in v]
plt.plot(v,z)
plt.ylim(-0.1,1.1)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

plt.title('50Hz-频谱')
# plt.savefig("D:\桌面\\1(6).jpg",dpi=300)
plt.show()
```

```python
delta=1/50
Sa=Piecewise((sin(t)/t,abs(t)>delta), (1, True))
x_recovery=summation(x.subs(t,n*delta)*Sa.subs(t,(t-n*delta)*pi/delta),(n,-20,20))
v=np.arange(-0.1,0.1+0.002,0.002)
v[abs(v)<1e-10]=0
z=[complex(x_recovery.subs(t,i).evalf()) for i in v]
plt.plot(v,z)
plt.ylim(-11,50)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.title('50Hz-恢复')
# plt.savefig("D:\桌面\\1(9).jpg",dpi=300)
plt.show()
```
