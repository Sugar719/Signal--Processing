# 任务四

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N =  150
x = [6*i**2+30 for i in range(150)]
X = np.zeros(N) + np.zeros(N)*1j # 频域频谱
# DTF变换
for k in range(N):
    for n in range(150):
        X[k] = X[k] + x[n]*np.exp(-1j*n*k/N*2*np.pi)

        
fig = plt.figure(figsize=(6, 6))
n1 = np.arange(0, 150, 1)
plt.plot(n1, x)
plt.title(u'x(n)')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 5))
n2 = np.arange(0, N, 1)
plt.stem(n2, abs(X))
plt.title(u'X(k)')
plt.show()
```

```python

fig = plt.figure(figsize=(10, 5))
n2 = np.arange(0, N, 1)
plt.stem(n2, np.angle(X, deg=True))
plt.title(u'DFT计算X(k)')
plt.show()
```

```python
# IDFT
x_p = np.zeros(len(x), dtype=complex)
for n in range(len(x)):
    for k in range(N):
        x_p[n] = x_p[n] + 1/N*X[k]*np.exp(1j*n*k/N*2*np.pi)

fig = plt.figure(figsize=(6, 6))
plt.plot(n1, x_p)
plt.title(u'IDFT计算x(n)')
plt.show()
```
